## 3.0 learning about loops

```python
# create a subset of the first million rows
df = ords_prods_merged[:1000000]
df.head()
```

```python
df.shape
```

```python
# define price_label function
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] >5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# apply it
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df.head()
```

```python
#get rid of the now unnecessary _merge column
df = df.drop(columns = '_merge')
```

```python
df.head()
```

```python
#dropping _merge from the big dataframe too
ords_prods_merged = ords_prods_merged.drop(columns = '_merge')
```

```python
ords_prods_merged.head()
```

```python
# see how many of each range we've found
df['price_range'].value_counts()
```

```python
# check the biggest price in my subset
df['prices'].max()
```

```python
#rewriting the if:then statement with .loc[if,then]
# high range
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# mid range
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
#low range
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#take a look at our new column
df['price_range_loc'].value_counts(dropna = False)
```

### making a column which denotes busiest day, regular days and least busy day of the week


```python
#busiest days of the week
ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

```python
# making a list of when it was a busy day
result = []

for value in ords_prods_merged['order_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('regularly busy')
```

```python
#looking at the list
result
```

```python
# make a column for the results inside my dataframe
ords_prods_merged['busiest_day'] = result
```

```python
# see how often my new labels show up
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## making a  column for 2 busiest days, regularly busy, or 2 least busy days

```python
#make a list within the new paramaters
results = []

for value in ords_prods_merged['order_day_of_week']:
    if (value == 0) or (value == 1):
        results.append('Busiest days')
    elif (value == 4) or (value == 3):
        results.append('Least busy')
    else:
        results.append('regularly busy')
```
